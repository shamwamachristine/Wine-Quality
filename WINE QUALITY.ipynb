```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
wine = pd.read_csv('/home/christine/Downloads/wine quality/winequality-red.csv')
```

```python
data = []
header = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']

# read in the data file
with open('/home/christine/Downloads/wine quality/winequality-red.csv') as f:
    # skip the first line since it contains the headers
    next(f)
    # loop over the remaining lines
    for line in f:
        # split the line into fields using the semicolon delimiter
        fields = line.strip().split(';')
        # convert the fields to their appropriate data types
        fixed_acidity = float(fields[0])
        volatile_acidity = float(fields[1])
        citric_acid = float(fields[2])
        residual_sugar = float(fields[3])
        chlorides = float(fields[4])
        free_sulfur_dioxide = float(fields[5])
        total_sulfur_dioxide = float(fields[6])
        density = float(fields[7])
        pH = float(fields[8])
        sulphates = float(fields[9])
        alcohol = float(fields[10])
        quality = int(fields[11])
        # create a dictionary to hold the sample data
        sample = {
            'fixed acidity': fixed_acidity,
            'volatile acidity': volatile_acidity,
            'citric acid': citric_acid,
            'residual sugar': residual_sugar,
            'chlorides': chlorides,
            'free sulfur dioxide': free_sulfur_dioxide,
            'total sulfur dioxide': total_sulfur_dioxide,
            'density': density,
            'pH': pH,
            'sulphates': sulphates,
            'alcohol': alcohol,
            'quality': quality
        }
        # add the sample to the data list
        data.append(sample)

# create a DataFrame from the data
wine = pd.DataFrame(data, columns=header)
```

## Question

Test the claim that Wines with a higher quality rating have a higher median alcohol content than wines with a lower quality rating.

* Ho = Median alcohol content of low quality rating = Median alcohol content of low quality rating wine 
* H1 = Median alcohol content of high quality rating wine > Median alcohol content of low quality wine

```python
wine_high_quality =wine[wine['quality']>=7]['alcohol']

wine_low_quality = wine[wine['quality']<7]['alcohol']


```

```python
wine_high_quality
```

```python
wine_low_quality
```

```python
t_statistics,p_value=ttest_ind(wine_high_quality,wine_low_quality)
```

```python
t_statistics,p_value
```

Conclusion: With the p_value higher than the threshold we hereby reject the null hypothesis and conclude that wines with higher rating have a high median alcohol content than wines with low quality rating.
